```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0.000
sigma = 0.005
start_price = 370 #Ethereum기준 370만시작?

np.random.seed(0)
returns = np.random.normal(loc=mu, scale=sigma, size = 30)
price = start_price*(1+returns).cumprod()

plt.plot(price)
```
